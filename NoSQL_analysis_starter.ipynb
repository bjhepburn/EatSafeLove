### 1. Which establishments have a hygiene score equal to 20?

```python
establishments.find_one({})
```

```python
# Find the establishments with a hygiene score of 20
query = {'scores.Hygiene': {'$eq' :20}}

# Use count_documents to display the number of documents in the result
hygiene20 = establishments.count_documents(query)
print(f'Number of establishments with hygiene score equal to 20: {hygiene20}')
# Display the first document in the results using pprint
pprint(establishments.find(query)[0])
```

```python
# Convert the result to a Pandas DataFrame
hygiene_df = pd.DataFrame(establishments.find(query))
# Display the number of rows in the DataFrame
print(len(hygiene_df))
# Display the first 10 rows of the DataFrame
hygiene_df.head(3)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query2 = {'$and':[{'LocalAuthorityName': {'$regex':'London'}},
                  {'RatingValue':{'$gte':'4'}}
                  ]}

# Use count_documents to display the number of documents in the result
print(establishments.count_documents(query2))
# Display the first document in the results using pprint
pprint(list(establishments.find(query2)))
```

```python
# Convert the result to a Pandas DataFrame
london_df = pd.DataFrame(establishments.find(query2))
# Display the number of rows in the DataFrame
print(len(london_df))
# Display the first 10 rows of the DataFrame
london_df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of '5', sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

degree_search = 0.01
latitude_low = establishments.find_one({'BusinessName':'Penang Flavours'})['geocode']['latitude'] - degree_search
latitude_high = establishments.find_one({'BusinessName':'Penang Flavours'})['geocode']['latitude'] + degree_search

longitude_low = establishments.find_one({'BusinessName':'Penang Flavours'})['geocode']['longitude'] - degree_search
longitude_high = establishments.find_one({'BusinessName':'Penang Flavours'})['geocode']['longitude'] + degree_search

query = {'$match': {'$and':
            [{'RatingValue': {'$eq':'5'}},
            {'geocode.latitude':{'$gte':latitude_low,'$lte':latitude_high}},
            {'geocode.longitude':{'$gte':longitude_low,'$lte':longitude_high}}]
            }}
sort =  {'$sort':{'count':-1}}
pipeline = [query,sort]

# Print the results
close_list = list(establishments.aggregate(pipeline))
for i in close_list:
    print(i)
```

```python
# Convert result to Pandas DataFrame
nearby_df = pd.DataFrame(close_list)
nearby_df.head()
```
